```python
import torch
```

```python
X = torch.arange(1,6)
X = X.reshape(5,1)
X = X.float()
X
```

```python
y = torch.arange(10, 15)
y = y.reshape(5,1)
```

```python
W = torch.rand(1, 5, requires_grad = True)
b = torch.rand(1, requires_grad = True)
W = W.float()
# b = b.float
W
```

```python
lr = 0.011
```

```python
for epoch in range(1001):

  y_pred = torch.matmul(X, W) + b

  y_pred = torch.relu(y_pred)

  loss = torch.sqrt(torch.mean((y_pred - y)) ** 2)

  loss.backward()

  with torch.no_grad():
    W -= lr * W.grad
    b -= lr * b.grad

    W.grad.zero_()
    b.grad.zero_()

  print(epoch, loss.item())

print("The final weights ", W)
print("The final bias ", b)
```
