```python
import torch
```

```python
x = torch.tensor([[1, 2, 3, 4],
 [5, 6, 7, 10]], dtype = float)
# x.float()
```

```python
y = torch.tensor([10, 12, 15, 17])
```

```python
# Weights and Bias

W1 = torch.randn(4,2, requires_grad = True, dtype = float)
b1 = torch.randn(1,1, requires_grad = True)

W2 = torch.randn(2, 1, requires_grad = True, dtype = float)
b2 = torch.randn(1,1, requires_grad = True)
```

```python
lr = 0.001

# Training Loop

for epoch in range(1000):

# Forward pass

  res = (x @ W1) + b1
  res = torch.relu(res)

  pred = (res @ W2) + b2

  # Loss
  # MSE
  loss = torch.mean((pred - y) ** 2)

  loss.backward()

  with torch.no_grad():
    W1 -= lr * W1.grad
    b1 -= lr * b1.grad

    W2 -= lr + W2.grad
    b2 -= lr + b2.grad

    W1.grad.zero_()
    W2.grad.zero_()
    b1.grad.zero_()
    b2.grad.zero_()

  print(epoch + 1, loss.item())


print("The final weights ", W1)
print("The final weights ", W2)
print("The final bias ", b1)
print("The final bias ", b2)
```
